```python
import ccxt

exchange = ccxt.binanceus()
exchange.load_markets()
symbol = 'BTC/USDT'
amount = 1.2345678  # amount in base currency BTC
price = 87654.321  # price in quote currency USDT
formatted_amount = exchange.amount_to_precision(symbol, amount)
formatted_price = exchange.price_to_precision(symbol, price)
print(formatted_amount, formatted_price)
```

```python
import ccxt

exchange = ccxt.binanceus()
symbol = 'BTC/USDT'  # Example symbol
ticker = exchange.fetch_ticker(symbol)
price = ticker['last']  # Get the last traded price
volume24h = ticker.baseVolume # this gives last 24 h volume for binance only
print(f"The price of {symbol} is {price}")
print(f"The volume of {symbol} is {volume24h}")
```

```python
import requests
from datetime import datetime

# docs: https://docs.bitfinex.com/reference/rest-public-ticker
url = "https://api-pub.bitfinex.com/v2/ticker/tBTCUSD"

headers = {"accept": "application/json"}

response = requests.get(url, headers=headers)
btc_data = response.json()

print(response.text)

keys = ['BID', 'BID_SIZE', 'ASK', 'ASK_SIZE', 'DAILY_CHANGE', 'DAILY_CHANGE_RELATIVE', 'LAST_PRICE', 'VOLUME', 'HIGH', 'LOW']

price = btc_data[keys.index('LAST_PRICE')]
volume = btc_data[keys.index('VOLUME')]
date_time = datetime.now()  # Assuming the current datetime for logging; adjust as necessary

symbol = 'BTC'
print(f"The price of {symbol} is {price}")
print(f"The volume of {symbol} is {volume}")
```
